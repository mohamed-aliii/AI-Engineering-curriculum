## The "Why"

Modern AI Engineering is not just API calls and prompt templates. Without a deep understanding of the following, the LLMOps and System Design tracks (Main Track 15+) will be **impossible to truly master**:

| Foundation | Why It's Non-Negotiable | Where It Appears in Main Track |
|------------|------------------------|-------------------------------|
| **Probability & Statistics** | Every model output is a probability distribution. You must understand what confidence, calibration, and likelihood mean. | NB15 (loss functions), NB37 (LLM evaluation), NB42 (system design) |
| **Linear Algebra & Optimization** | Gradient descent is the engine of all training. LoRA adapters are literally low-rank matrix decompositions. | NB15 (QLoRA fine-tuning), NB29 (inference optimization) |
| **Classical Models** | Logistic Regression, SVMs, Decision Trees, and Ensembles are still the correct choice for 80% of production ML. A senior engineer reaches for XGBoost before GPT-4. | NB15 (model selection), NB42 (system design interviews) |
| **Evaluation Metrics** | Precision, Recall, F1, AUC-ROC, calibration curves — if you can't measure a model, you can't improve it or trust it in production. | NB32 (testing), NB35 (monitoring), NB37 (evaluation) |
| **Bias-Variance Trade-off** | Overfitting, underfitting, regularization, and cross-validation govern every training decision you will ever make. | NB15 (training), NB35 (drift detection), NB42 (design) |
| **Bayesian Thinking** | Prior beliefs, posterior updates, and uncertainty quantification are essential for responsible AI and guardrails. | NB40 (AI safety), NB41 (governance) |

**The blunt truth:** An engineer who skips these foundations will be able to *copy-paste* fine-tuning code, but will be **completely lost** when a model fails in production, when an interviewer asks *why*, or when a system design requires choosing between a Random Forest and a Transformer.

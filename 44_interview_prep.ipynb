## Part 1: ML Fundamentals (Questions 1-15)

### 🧠 Q1: What is the bias-variance tradeoff?
<details><summary>Answer (NB13)</summary>

**Bias** = error from overly simplistic model assumptions (underfitting). **Variance** = error from sensitivity to training data fluctuations (overfitting). The tradeoff: simple models have high bias/low variance; complex models have low bias/high variance. The optimal model minimizes total error = bias² + variance + irreducible noise. Regularization (L1/L2) controls this tradeoff.
</details>

### 🧠 Q2: Explain the difference between L1 and L2 regularization.
<details><summary>Answer (ML/02)</summary>

L1 (Lasso) adds |w| penalty → drives weights exactly to zero → automatic feature selection. L2 (Ridge) adds w² penalty → shrinks weights toward zero but never exactly zero. Geometrically: L1 constraint is a diamond (sharp corners on axes = sparsity), L2 is a circle (smooth = no sparsity).
</details>

### 🧠 Q3: What is gradient descent, and why do we use mini-batch GD?
<details><summary>Answer (NB13, ML/01, DL/04)</summary>

GD iteratively adjusts model parameters in the direction of steepest loss decrease. **Batch GD**: uses all training data per step (accurate but slow). **Stochastic GD**: uses 1 sample (fast but noisy). **Mini-batch GD**: uses 32-512 samples (best of both — GPU-efficient, good convergence). Mini-batch also provides regularization through noise.
</details>

### 🧠 Q4: How do you handle class imbalance?
<details><summary>Answer (NB13, 03b_eda)</summary>

Five strategies: (1) Class weights in loss function, (2) SMOTE/oversampling minority class, (3) Undersampling majority class, (4) Change metric from accuracy to F1/AUROC/precision@k, (5) Ensemble methods (balanced random forests). In production fraud detection, use precision@recall=0.95 as the metric.
</details>

### 🧠 Q5: What is cross-validation and why is it important?
<details><summary>Answer (NB13, ML/06)</summary>

K-fold CV splits data into K parts, trains on K-1, validates on 1, rotates K times. Gives a more reliable performance estimate than a single train/test split. Use stratified K-fold for classification (preserves class ratios). Time-series data requires temporal splits (no random shuffling — future can't train past).
</details>

### 💻 Q6: Explain the attention mechanism.
<details><summary>Answer (DL/07, NLP/04)</summary>

Attention(Q,K,V) = softmax(QK^T/√d_k)V. Each token computes a Query (what am I looking for?), every token has a Key (what do I contain?) and Value (my actual content). Dot-product Q·K gives relevance scores, softmax normalizes to probabilities, multiply by V gives weighted output. Multi-head attention runs this in parallel with different projections to capture syntactic, semantic, and positional relationships simultaneously.
</details>

### 💻 Q7: What is backpropagation?
<details><summary>Answer (DL/01, 10b_pytorch)</summary>

Backpropagation is the chain rule applied recursively through a computational graph. Forward pass: compute output and loss. Backward pass: compute ∂Loss/∂w for every weight by multiplying local gradients backward through the graph. PyTorch's autograd does this automatically — `.backward()` traverses the graph in reverse topological order, calling each node's `_backward()` function.
</details>

### 🧠 Q8: What is the vanishing gradient problem and how is it solved?
<details><summary>Answer (NLP/03, DL/02, DL/03)</summary>

In deep networks, gradients are multiplied through many layers. If gradients < 1, they exponentially shrink to zero → early layers stop learning. Solutions: (1) ReLU activation (gradient is 1 for positive inputs), (2) Residual connections (gradient highway bypassing layers), (3) Batch/layer normalization, (4) Better initialization (Xavier/He), (5) LSTM gates (for recurrent networks).
</details>

### 🧠 Q9: Explain the transformer architecture.
<details><summary>Answer (DL/07)</summary>

A stack of identical blocks, each containing: (1) Multi-Head Self-Attention + residual + LayerNorm, (2) Feed-Forward Network + residual + LayerNorm. Encoder (BERT) uses bidirectional attention. Decoder (GPT) uses causal masking. Key innovation over RNNs: processes all tokens in parallel (O(1) depth) instead of sequentially (O(n)). Positional encoding (sinusoidal or RoPE) injects position information.
</details>

### 🧠 Q10: What is the difference between pre-training and fine-tuning?
<details><summary>Answer (NLP/05, NB15)</summary>

Pre-training: Train on massive unlabeled data with self-supervised objective (MLM for BERT, CLM for GPT). Learns general language understanding. Fine-tuning: Adapt the pre-trained model to a specific task with labeled data. Uses much lower learning rate (2e-5 vs 1e-3) to preserve pre-trained knowledge. LoRA (NB15) is parameter-efficient fine-tuning that adds small adapters instead of updating all weights.
</details>

### 🤖 Q11: Explain LoRA and why it's used.
<details><summary>Answer (NB15)</summary>

LoRA (Low-Rank Adaptation) freezes the pre-trained model and adds small rank-r matrices (A and B) to attention projections: W_new = W_frozen + A·B. Instead of updating 7B parameters, you train ~0.1% (7M). This reduces GPU memory from 28GB (full fine-tune at fp16) to ~6GB (QLoRA at 4-bit). Target modules are typically q_proj, k_proj, v_proj — the Q/K/V projection matrices in multi-head attention.
</details>

### 🤖 Q12: What is RAG and when would you use it vs fine-tuning?
<details><summary>Answer (NB21)</summary>

RAG retrieves relevant documents and injects them into the prompt context. Use RAG when: data changes frequently, you need citations/sources, proprietary data that can't be baked into model weights. Use fine-tuning when: you need to change the model's behavior/style, domain-specific terminology, structured output formats. Many production systems use BOTH: fine-tune for style, RAG for facts.
</details>

### 🤖 Q13: What are embeddings and how do they work?
<details><summary>Answer (NLP/02, NB20)</summary>

Embeddings map discrete tokens/documents to continuous vectors where semantic similarity = geometric proximity. Word2Vec learns from word co-occurrence. BERT produces contextual embeddings (same word gets different vectors based on context). For RAG (NB21), sentence-transformers encode entire paragraphs — similar documents have high cosine similarity. Stored in vector databases (ChromaDB, pgvector, NB20) for fast retrieval.
</details>

### 🤖 Q14: Explain quantization and its tradeoffs.
<details><summary>Answer (NB29)</summary>

Quantization reduces model precision: float32 (4 bytes) → float16 (2 bytes) → int8 (1 byte) → int4 (0.5 bytes). A 7B model goes from 28GB (fp32) to 3.5GB (4-bit). Tradeoffs: lower precision = slightly lower quality. GPTQ: post-training quantization, good quality. AWQ: activation-aware, better quality. GGUF: CPU-optimized format (llama.cpp). BitsAndBytes: dynamic quantization for training (QLoRA).
</details>

### 🤖 Q15: What is prompt injection and how do you defend against it?
<details><summary>Answer (NB40)</summary>

Prompt injection: user input tricks the LLM into ignoring system instructions ("ignore all previous instructions and..."). Defenses: (1) Input sanitization, (2) System prompt hardening, (3) Output validation (check for policy violations), (4) NeMo Guardrails / Guardrails AI for topic containment, (5) Separate user input from instructions with delimiters and structural prompting.
</details>

## Part 2: MLOps & Infrastructure (Questions 16-30)

### ⚙️ Q16: Explain Docker and why it's used for ML.
<details><summary>Answer (NB02)</summary>

Docker packages your model + dependencies + runtime into an isolated container. Solves "works on my machine" — same container runs identically in dev, staging, production. For ML: pin exact Python version, CUDA version, library versions. Docker image → pushed to registry → deployed to Kubernetes (NB02) for scaling.
</details>

### ⚙️ Q17: What is a feature store and why use it?
<details><summary>Answer (NB11)</summary>

A feature store (Feast) serves pre-computed features at inference time. Solves: (1) Training-serving skew (same feature computation logic), (2) Latency (pre-computed aggregations served in <10ms), (3) Feature sharing (teams reuse features). Without it: each model computes its own features, leading to inconsistencies and duplicated work.
</details>

### ⚙️ Q18: What is model drift and how do you detect it?
<details><summary>Answer (NB35)</summary>

Model drift: model performance degrades over time because production data distribution changes. Types: data drift (input distribution changes), concept drift (relationship between features and target changes). Detection: EvidentlyAI computes statistical tests (KS test, PSI) comparing training vs production distributions. Alert when p-value < threshold. Remediation: retrain on recent data.
</details>

### ⚙️ Q19: Explain CI/CD for ML models.
<details><summary>Answer (NB31)</summary>

ML CI/CD extends software CI/CD with: (1) Data validation tests, (2) Model quality gates (accuracy must exceed threshold), (3) Model registry (version control for models, MLflow), (4) Staged rollout (canary deployment: 5% traffic → monitor → 100%). GitHub Actions triggers on code push, Kubeflow Pipelines orchestrates training → evaluation → deployment.
</details>

### ⚙️ Q20: What is Kubernetes and why is it important for ML?
<details><summary>Answer (NB02)</summary>

K8s orchestrates containers across a cluster. For ML: auto-scales model servers based on traffic (HPA), manages GPU node pools, handles rolling deployments (zero-downtime updates), and provides service discovery. A typical production setup: model served via FastAPI in Docker → deployed as K8s Deployment → exposed via Service → auto-scaled by HPA.
</details>

### 🤖 Q23: What is vLLM and PagedAttention?
<details><summary>Answer (NB29)</summary>

vLLM is an inference engine for LLMs. PagedAttention manages KV-cache memory like OS virtual memory — allocates non-contiguous memory blocks instead of pre-allocating maximum sequence length. This eliminates memory waste (up to 60-80% was wasted in naive approaches). Combined with continuous batching (new requests added mid-generation), vLLM achieves 5-24x higher throughput than HuggingFace's default.
</details>

### 🤖 Q24: What is LangGraph and when would you use it?
<details><summary>Answer (NB23)</summary>

LangGraph models agent workflows as state machines (graphs). Nodes = processing steps, edges = transitions, state = shared context. Use over simple chains when: (1) workflow has loops/retries, (2) multiple agents collaborate, (3) you need checkpointing/human-in-the-loop, (4) deterministic control flow around stochastic LLM reasoning. Key principle: orchestrator (deterministic) controls flow, LLM (stochastic) provides reasoning.
</details>

### 🤖 Q25: How would you evaluate an LLM-based system?
<details><summary>Answer (NB37)</summary>

Multi-layer evaluation: (1) Component-level: retrieval recall/precision (RAG), tool accuracy (agents), (2) End-to-end: DeepEval metrics (faithfulness, relevancy, hallucination), LLM-as-a-Judge, (3) Human evaluation: preference ranking, task completion rate. For RAG: RAGAS metrics (context precision, context recall, faithfulness, answer relevancy). Use Langfuse for tracing individual LLM calls to root-cause failures.
</details>

### 🏗️ Q26: What is the difference between batch and real-time ML?
<details><summary>Answer (NB10, NB11, NB42)</summary>

Batch: process data periodically (hourly/daily). Good for: recommendations, reports, retraining. Tech: Spark (NB08), Airflow. Real-time: process events as they arrive (<100ms). Good for: fraud, search ranking, notifications. Tech: Kafka + Flink (NB10), Feature Store (NB11). Many systems use both: batch for training features, real-time for serving features.
</details>

### 💻 Q27: What is async/await and why is it important for AI APIs?
<details><summary>Answer (NB01, NB27)</summary>

Async allows a server to handle thousands of concurrent requests without blocking. While waiting for an LLM API response (~2s), the server handles other requests. FastAPI with async (NB27) can handle 10x more concurrent LLM requests than synchronous code. Key pattern: `async def predict(request)` + `await litellm.acompletion(...)`. Essential for production AI APIs.
</details>

### ⚙️ Q28: Explain SHAP values.
<details><summary>Answer (NB41)</summary>

SHAP (SHapley Additive exPlanations) attributes each feature's contribution to a specific prediction. Based on game theory: each feature is a "player" and SHAP computes each player's fair contribution to the prediction. Positive SHAP = pushes prediction higher, negative = lower. Required for: regulatory compliance (explain why loan was denied), debugging (why did model misclassify?), trust (show stakeholders what the model relies on).
</details>

### 🤖 Q29: What is DPO and how does it relate to RLHF?
<details><summary>Answer (NB16)</summary>

Both align LLMs to human preferences. RLHF: train a reward model on preference data → use PPO to optimize the LLM against the reward model (complex, unstable). DPO (Direct Preference Optimization): reformulates the RLHF objective to directly optimize on preference pairs without a reward model (simpler, more stable, same results). DPO needs pairs of (chosen, rejected) responses for the same prompt.
</details>

### ⚙️ Q30: How do you monitor an LLM in production?
<details><summary>Answer (NB36, NB37)</summary>

Four pillars: (1) Latency (p50/p95/p99 per request), (2) Cost (token usage per request, daily spend), (3) Quality (LLM-as-judge scoring, user feedback), (4) Safety (toxicity rate, policy violations). Tools: Langfuse for tracing (see exactly which LLM call failed), token budgeting to prevent runaway costs, automated quality regression tests (DeepEval, NB37).
</details>

### 💻 Q31: Write a PyTorch training loop.
<details><summary>Answer (10b_pytorch)</summary>

```python
for epoch in range(epochs):
    model.train()
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()              # 1. Clear gradients
        predictions = model(X_batch)       # 2. Forward pass
        loss = loss_fn(predictions, y_batch) # 3. Compute loss
        loss.backward()                    # 4. Backward pass
        optimizer.step()                   # 5. Update weights
```
The 5 sacred steps. `model.train()` enables dropout. `model.eval()` + `torch.no_grad()` for evaluation.

### 💻 Q32: Implement scaled dot-product attention.
<details><summary>Answer (DL/07)</summary>

```python
def attention(Q, K, V, mask=None):
    d_k = Q.shape[-1]
    scores = Q @ K.transpose(-2, -1) / math.sqrt(d_k)
    if mask is not None:
        scores = scores.masked_fill(mask == 0, -1e9)
    weights = F.softmax(scores, dim=-1)
    return weights @ V
```
</details>

### 💻 Q33: How do you implement BPE tokenization?
<details><summary>Answer (NLP/01)</summary>

1. Start with character vocabulary. 2. Count all adjacent character pairs. 3. Merge the most frequent pair into one token. 4. Repeat N times. Result: common words are single tokens ("the"), rare words are split into learned subwords ("un"+"happi"+"ness"). GPT-4 uses ~100K BPE tokens.
</details>

### 💻 Q35: How do you handle GPU out-of-memory errors during training?
<details><summary>Answer (10b_pytorch, NB15, NB29)</summary>

Ordered by impact: (1) Reduce batch size, (2) Use gradient accumulation (simulate larger batch), (3) Use mixed precision (bf16/fp16), (4) Use gradient checkpointing (recompute vs store activations), (5) Use QLoRA (4-bit base model), (6) Use DeepSpeed ZeRO (shard model across GPUs). For inference: quantize model (NB29), use flash attention, reduce max sequence length.
</details>
